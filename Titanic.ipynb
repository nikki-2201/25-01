```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("train.csv")
# df.isna().sum()
df.head()
```

```python
df.isna().sum()
```

```python
sns.barplot(data=df,x="Pclass",y="Survived",errorbar=None,palette="viridis_r")
plt.xlabel("Pclass")
plt.ylabel("Survived")
plt.title("Bar plot")
plt.show()
```

```python
sns.scatterplot(data=df,x="Age",y="Fare",hue="Sex")
plt.xlabel("Age")
plt.ylabel("Fare")
plt.title("Scatter plot")
plt.show()
```

```python
s=df[['Age', 'Fare', 'Pclass','Survived']]
sns.pairplot(s,hue='Survived',diag_kind='kde',palette="crest")
plt.title("relation between age fare and p class",x=-1,y=3.3)
plt.show()
```

```python
df_numeric = df.select_dtypes(include=['number'])
s = df_numeric.corr()
sns.heatmap(s, annot=True, cmap="PuBuGn", fmt=',.2f', linewidths=0.5)
plt.title("Correlation Heatmap")
plt.show()
```
