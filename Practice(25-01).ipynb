```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("GlobalLandTemperaturesByCity.csv")
```

```python
df.head()
```

```python
import matplotlib.pyplot as plt
c=df[df['City'] == 'New York'].copy()
c['dt']=pd.to_datetime(c['dt'],errors='coerce')
c['Year']=c['dt'].dt.year
annual_avg_temp=c.groupby('Year')['AverageTemperature'].mean().reset_index()
plt.figure(figsize=(12,6))
plt.plot(annual_avg_temp['Year'],annual_avg_temp['AverageTemperature'],linestyle='-',color='purple',marker="d")
plt.title("Average temperature trend in New York",fontsize=16)
plt.xlabel('Year',fontsize=12)
plt.ylabel('Average temperature',fontsize=12)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv("GlobalLandTemperaturesByCity.csv")
df.dropna(subset=['AverageTemperature'], inplace=True)
avg_temp_by_country = df.groupby('Country')['AverageTemperature'].mean()
sorted_avg_temp = avg_temp_by_country.sort_values(ascending=False)
top_10_countries = sorted_avg_temp.head(10)
#bar plot
plt.figure(figsize=(12, 6))
plt.bar(top_10_countries.index, top_10_countries.values,color="green")
plt.xlabel("Country")
plt.ylabel("Average Temperature")
plt.title("Top 10 Countries with Highest Average Temperatures")
plt.xticks(rotation=45, ha='right') 
plt.show()
print("Top 10 Countries with Highest Average Temperatures:")
top_10_countries
```

```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv("GlobalLandTemperaturesByCity.csv")
df['AverageTemperature']=df['AverageTemperature'].fillna(df['AverageTemperature'].mean())
df['AverageTemperature']
plt.hist(df['AverageTemperature'],bins=50,edgecolor="black")
plt.xlabel("Temparature",fontsize=12)
plt.ylabel("Frequency",fontsize=12)
plt.grid(axis="y",linestyle="--",alpha=0.7)
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd


df['AverageTemperature'] = df['AverageTemperature'].fillna(df['AverageTemperature'].mean())


df['Year'] = pd.to_datetime(df['dt']).dt.year

country_df = df[df['Country'] == 'India']
temperature_by_year = [country_df[country_df['Year'] == year]['AverageTemperature'].values for year in country_df['Year'].unique()]

plt.boxplot(temperature_by_year, labels=country_df['Year'].unique())

plt.xlabel("Year")
plt.ylabel("Average Temperature")


plt.xticks(rotation=90)

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

c = df[df['Country'] == 'India'].copy()
c['dt'] = pd.to_datetime(c['dt'], errors='coerce')
c = c.dropna(subset=['dt'])
c['Year'] = c['dt'].dt.year
c = c[['Year', 'AverageTemperature']].dropna()

# Group the temperatures by year
g = [group['AverageTemperature'] for _, group in c.groupby('Year')]

# Create the boxplot
plt.figure(figsize=(12,6))
plt.boxplot(g, showfliers=True)  
plt.title("Temperature spread in India")
plt.xlabel('Year')
plt.ylabel('Average Temperature (°C)')
plt.show()
```
